```python
# Import Libraries Required
import pandas as pd
```

```python
# Load in the data that needs to be cleaned
clients_list = pd.read_csv("Resources/Sales_Report_07_06.csv")

#Read the data
clients_list
```

```python
# Drop rows that are empty in this case anything that is empty in email
clients_list = clients_list.dropna(axis=0, subset=['Email'])

# View Results
clients_list
```

```python
# Drop a row based on a certain string it contains
# df[df["column name"].str.contains("string to delete")==False]
final_clients_list = clients_list[clients_list["Email"].str.contains("Email")==False]

# Show the list
final_clients_list
```

```python
# Grab the columns that you want only
download_list = final_clients_list[['Name', 'Phone', 'Email', 'Location']].set_index('Name')

# Show the list
download_list
```

```python
# Download to Resources and name whatever you want after the '/'
download_list.to_csv('Resources/full_client_list_0706.csv')
```
